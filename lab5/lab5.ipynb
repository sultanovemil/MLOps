```python
# Загрузка необходимых библтотек
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Создание качественных датасетов

# Качественный датасет 1
X_quality_1 = np.random.rand(100, 1)
y_quality_1 = 2 * X_quality_1.squeeze() + np.random.normal(0, 0.15, 100)

# Качественный датасет 2
X_quality_2 = np.random.rand(100, 1)
y_quality_2 = 2 * X_quality_2.squeeze() + np.random.normal(0, 0.20, 100)

# Качественный датасет 3
X_quality_3 = np.random.rand(100, 1)
y_quality_3 = 2 * X_quality_3.squeeze() + np.random.normal(0, 0.25, 100)
```

```python
# Визуализация отношения между входным признаком (X_quality) и целевой переменной (y_quality)
plt.scatter(X_quality_1, y_quality_1)
plt.scatter(X_quality_2, y_quality_2)
plt.scatter(X_quality_3, y_quality_3)
plt.show()
```

```python
# Обучение модели линейной регрессии на качественном наборе данных №1
X_train, X_test, y_train, y_test = train_test_split(X_quality_1, y_quality_1, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Создание набора данных с шумом
X_noisy = np.random.rand(100, 1)
y_noisy = 2 * X_noisy.squeeze() + np.random.normal(0, 0.5, 100)
```

```python
# Визуализация отношения между входным признаком (X_noisy) и целевой переменной (y_noisy)
plt.scatter(X_noisy, y_noisy)
plt.show()
```

```python
# Тестирование модели
y_pred_quality_2 = model.predict(X_quality_2)
y_pred_quality_3 = model.predict(X_quality_3)
y_pred_noisy = model.predict(X_noisy)
```

```python
mse_2 = mean_squared_error(y_quality_2, y_pred_quality_2)
mse_3 = mean_squared_error(y_quality_3, y_pred_quality_3)
mse_noise = mean_squared_error(y_noisy, y_pred_noisy)
print(f"MSE на качественном наборе данных №2 {mse_2}")
print(f"MSE на качественном наборе данных №3 {mse_3}")
print(f"MSE на наборе данных с шумом {mse_noise}")
```

```python
# Анализ с помощью Pytest
def test_quality_predictions_2():
    assert mean_squared_error(y_quality_2, y_pred_quality_2) < 0.1, "Производительность модели на качественном наборе данных 2 неудовлетворительна"

def test_quality_predictions_3():
    assert mean_squared_error(y_quality_3, y_pred_quality_3) < 0.1, "Производительность модели на качественном наборе данных 3 неудовлетворительна"

def test_noisy_predictions():
   assert mean_squared_error(y_noisy, y_pred_noisy) < 0.1, "Модель имеет проблемы с шумным набором данных"
```

```python
%%writefile sample_data/app.py
# Загрузка необходимых библтотек
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Создание качественных датасетов

# Качественный датасет 1
X_quality_1 = np.random.rand(100, 1)
y_quality_1 = 2 * X_quality_1.squeeze() + np.random.normal(0, 0.15, 100)

# Качественный датасет 2
X_quality_2 = np.random.rand(100, 1)
y_quality_2 = 2 * X_quality_2.squeeze() + np.random.normal(0, 0.20, 100)

# Качественный датасет 3
X_quality_3 = np.random.rand(100, 1)
y_quality_3 = 2 * X_quality_3.squeeze() + np.random.normal(0, 0.25, 100)

# Обучение модели линейной регрессии на качественном наборе данных
X_train, X_test, y_train, y_test = train_test_split(X_quality_1, y_quality_1, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)

# Создание набора данных с шумом
X_noisy = np.random.rand(100, 1)
y_noisy = 2 * X_noisy.squeeze() + np.random.normal(0, 0.5, 100)

# Тестирование модели
y_pred_quality_2 = model.predict(X_quality_2)
y_pred_quality_3 = model.predict(X_quality_3)
y_pred_noisy = model.predict(X_noisy)

# Анализ с помощью Pytest
def test_quality_predictions_2():
    assert mean_squared_error(y_quality_2, y_pred_quality_2) < 0.1, "Производительность модели на качественном наборе данных 2 неудовлетворительна"

def test_quality_predictions_3():
    assert mean_squared_error(y_quality_3, y_pred_quality_3) < 0.1, "Производительность модели на качественном наборе данных 3 неудовлетворительна"

def test_noisy_predictions():
   assert mean_squared_error(y_noisy, y_pred_noisy) < 0.1, "Модель имеет проблемы с шумным набором данных"
```
